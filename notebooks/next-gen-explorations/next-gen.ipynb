# Renderings

Given a ticker, lets look at yfinance bar charts for 5 years, daily, weekly, monthly, yearly, quarterly

```python
ticker="TSLA"
data = pd.DataFrame(yf.download(ticker, interval="1d"))

moving_averages_days = [7, 30, 90, 180, 360]

def plot(plt_data):
    fig = go.Figure()
    for ma in moving_averages_days:
        fig.add_trace(go.Scatter(x=plt_data.index, y=plt_data[f'Close_diff_MA_{ma}'], mode="lines", name=f'Close_diff_MA_{ma}'))
    fig.show()

def plot_main(plt_data):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=plt_data.index, y=plt_data[f'Close'], mode="lines", name=f'Close'))
    fig.show()

data = data[['Close', 'Volume']]

for ma in moving_averages_days:
    data[f'Close_diff_MA_{ma}'] = data['Close'] - data['Close'].rolling(window=ma).mean()
    data[f'Volume_diff_MA_{ma}'] = data['Volume'] - data['Volume'].rolling(window=ma).mean()

plot(data)
plot_main(data)

```
